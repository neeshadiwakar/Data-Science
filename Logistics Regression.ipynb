```python
import numpy as np
data_new=data.replace("?",np.nan)
```

```python
data_new.isnull().sum()
```

```python
data_new["workclass"].fillna(data_new["workclass"].mode()[0],inplace=True)
data_new["occupation"].fillna(data_new["occupation"].mode()[0],inplace=True)
data_new["native-country"].fillna(data_new["native-country"].mode()[0],inplace=True)
```

```python
data_new.isnull().sum()
```

```python
colname=[]
for x in data_new.columns:
    if data_new[x].dtypes=='object':
        colname.append(x)
colname
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
for x in colname:
    data_new[x]=le.fit_transform(data_new[x])
    le_name_mapping = dict(zip(le.classes_,le.transform(le.classes_)))
    print('Feature',x)
    print('mapping',le_name_mapping)
```

```python
data_new
```

```python
x=data_new.values[:, 0:-1]
y=data_new.values[:,-1] 
```

```python
y
```

```python
x.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=10)
```

```python
print("shape of x-train",x_train.shape)
print("shape of x-test",x_test.shape)
print("shape of y-train",y_train.shape)
print("shape of y-test",y_test.shape)
```

```python
from sklearn .linear_model import LogisticRegression
classifier = LogisticRegression()
classifier.fit(x_train,y_train)
```

```python
y_pred=classifier.predict(x_test)
print(y_pred)
print(list(zip(y_test,y_pred)))
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
cfm=confusion_matrix(y_test,y_pred)
print(cfm)
print("classification Report:")
print(classification_report(y_test,y_pred))

acc=accuracy_score(y_test,y_pred)
print("Accuracy of the model :",acc)
```
